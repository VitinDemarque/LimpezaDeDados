```python
import unicodedata
```

```python
def normalizar_texto(texto):
    if pd.isnull(texto):
        return texto
    texto = str(texto).lower()
    texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('utf-8')
    return texto.strip()
```

```python
df['produto'] = df['produto'].apply(normalizar_texto)
df['marca'] = df['marca'].apply(normalizar_texto)
```

```python
import re
```

```python
def normalizar_produto(texto):
    if pd.isnull(texto):
        return texto
    texto = str(texto).lower()
    texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('utf-8')
    texto = re.sub(r'[^a-z0-9\s]', '', texto)
    return texto.strip()
```

```python
df['produto'] = df['produto'].apply(normalizar_produto)
```

```python
print(df['produto'].unique())
```

```python
correcoes = {
    'refrigkrante': 'refrigerante',
    'scl': 'sal',
    'agua mineras': 'agua mineral',
    'presuntd': 'presunto'
}

df['produto'] = df['produto'].replace(correcoes)
```

```python
print(sorted(df['produto'].unique()))
```

```python
def normalizar_marca(texto):
    if pd.isnull(texto):
        return texto
    texto = str(texto).lower()
    texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('utf-8')
    texto = texto.replace('-', ' ')
    texto = re.sub(r'[^a-z0-9\s]', '', texto)
    return texto.strip()

df['marca'] = df['marca'].apply(normalizar_marca)
```
